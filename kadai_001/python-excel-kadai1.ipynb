```python
import openpyxl
from datetime import datetime

# 新しいワークブックを作成
workbook = openpyxl.Workbook()
sheet = workbook.active

# 表のデータ記入
sheet['B2'] = '請求書'

sheet['B4'] = '株式会社ABC'
sheet['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'] = '担当者名:鈴木一郎 様'

sheet['F4'] = 'No'
sheet['G4'] = '0001'
sheet['F5'] = '日付'

today = datetime.today()
sheet['G5'] = today.strftime('%Y/%m/%d')

#表を作成
data = [
    ['商品名','商品A','商品B'],
    ['数量',2,1],
    ['単価',10000,15000],
    ['金額',20000,15000]
]

start_row = 10  # B10セルからスタート
start_col = 2  # B列

for i, row in enumerate(data):
    for j, value in enumerate(row):
        if i > 0 and j > 0:  # 商品名以外のセルについて
            sheet.cell(row=start_row + j, column= start_col + i).value = int(value)
        else:
            sheet.cell(row=start_row + j, column= start_col + i).value = value

#合計金額の計算と記入
sheet['E13'].value = sheet['E11'].value + sheet['E12'].value

sheet['B15'] = '小計'
sheet['B16'] = '消費税'
sheet['B17'] = '合計'

sheet['E15'].value = sheet['E13'].value
sheet['E16'].value = sheet['E13'].value*0.1
sheet['E17'].value = sheet['E15'].value + sheet['E16'].value

#日付のtodayオブジェクトを特定のフォーマットで文字列に変換する
backup_filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"
workbook.save(backup_filename)
```
