```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df = pd.DataFrame()

# ファイルを統合
for file_name in file_list:
    temp_df = pd.read_excel(file_name, skiprows=0)
    temp_df.columns = ['売上年', '商品', '金額（千円）']

    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df], ignore_index=True)

#print(df)

# 商品、売上年ごとに金額を合計
df_grouped = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()

print(df_grouped)

# エクセルへの保存
writer = pd.ExcelWriter("売上集計表.xlsx")
df_grouped.to_excel(writer, sheet_name="Sheet1", index=False)
writer.close()

#作成したexcelの整形と保存
workbook = openpyxl.load_workbook("売上集計表.xlsx")
sheet = workbook.active
change_color = sheet["A1:C1"]
fill = PatternFill(patternType="solid", fgColor="F2F2F2")

for row in change_color:
    for cell in row:
        cell.fill = fill

workbook.save("売上集計表.xlsx")
```
